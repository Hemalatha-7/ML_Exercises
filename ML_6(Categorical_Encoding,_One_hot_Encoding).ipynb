```python
#It refers to the process of converting categorical or textual data into
#numerical format, so that it can be used as input for algorithms to process.
#The reason for encoding is that most machine learning algorithms work with
#numbers and not with text or categorical variables
#One-Hot Encoding is another popular technique for treating categorical variables.
#It simply creates additional features based on the number of unique values in the categorical feature.
#Every unique value in the category will be added as a feature.
#One-Hot Encoding is the process of creating dummy variables
#Importing libraries and reading dataset
import pandas as pd
df=pd.read_csv('/content/house-prices.csv')
df
```

```python
#dummy variables
#In this, we have four categories so had four columns for four different dummy variables.
dummies=pd.get_dummies(df.Bedrooms)
dummies
```

```python
# Merging dummy variables
merged=pd.concat([df,dummies],axis='columns')
merged
```

```python
#Dropping â€˜Homeâ€™ column
final = merged.drop(['Home'], axis=1)
final
```

```python
#Dropping one dummy(2) column
final = final.drop([2], axis=1)
final
```

```python
#dropping multiple variables
X = final.drop(['Brick', 'Neighborhood'], axis=1)
X
```

```python
#Dependent variable
y = final.Price
```

```python
#Importing Linear Regression model
from sklearn.linear_model import LinearRegression
model = LinearRegression()
```

```python
# Fitting the model,Predicting the house price,Checking the accuracy
X.columns=X.columns.astype(str)
model.fit(X,y)
model.predict(X)
model.score(X,y)
```
